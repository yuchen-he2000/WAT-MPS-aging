```python
def log2FC_calculate(df, class_list):
    # Separate the df based on annotations
    df_0 = df.loc[:, [df.columns[i] for i in range(len(df.columns)) if class_list[i] == '0']]
    df_1 = df.loc[:, [df.columns[i] for i in range(len(df.columns)) if class_list[i] == '1']]
    
    average_df_0 = df_0.mean(axis=1)
    average_df_1 = df_1.mean(axis=1)
    
    FC = average_df_1.div(average_df_0.where(average_df_0 != 0, np.nan))
    log2FC = np.log2(FC.astype('float').where(FC != 0, np.nan))
    return log2FC
```

```python
def quick_paired_PLSDA(counts_df, metadata, gene_info, design_factor, exp_condition, ref_condition, savedir):
    """
    Performs PLSDA and plot PLSDA results with scatter plot. Exports major contributors of each latent variables to excel.
    
    """
    
    # Find the samples that will not be used in this comparison
    drop_indexes = metadata.index[~metadata[design_factor].isin([exp_condition, ref_condition])].tolist()
    
    # Drop the not needed samples from counts_df and metadata
    counts_df_drop = counts_df.drop(drop_indexes, axis = 1)
    metadata_drop = metadata.drop(drop_indexes, axis=0)
    print(metadata_drop)
    
    ## Prepare counts_df_drop for PLS-DA
    selected_sample_list = counts_df_drop.columns #Retrieve a list of counts_df_drop column names
    
    # Annotate ref as class 0 and exp as class 1
    class_list = []
    for i in selected_sample_list:
        if metadata_drop.loc[i,design_factor] == ref_condition:
            class_list.append('0') # ref_condition is class 0
        elif metadata_drop.loc[i,design_factor] == exp_condition:
            class_list.append('1') # exp_condition is class 1
        else:
            print('error: third class exist!')
            return
    
    # Annotate each sample in counts_df_drop
    temp_index = len(counts_df_drop)
    counts_df_drop.loc[temp_index] = class_list
    counts_df_drop = counts_df_drop.rename(index={temp_index: design_factor})
    
    # use PLSRegression class from scikit-learn to implement PLS-DA (Partial Least Squares Discriminant Analysis)
    from sklearn.cross_decomposition import PLSRegression
    from sklearn.model_selection import train_test_split
    
    # Separate features and labels
    X = counts_df_drop.drop(design_factor, axis=0)  #Features
    y = counts_df_drop.loc[design_factor] #Labels

    print(f'features:{len(X)}; target:{len(y)}')
    
    # PLS-DA 
    plsda = PLSRegression(n_components=4)  # Choose the number of components
    plsda.fit(X.T, y)
    
    # Extract loadings for each latent variables for future analysis
    loadings = plsda.x_loadings_

    # Convert the loadings for each feature to a dataframe
    LV_list = ['LV1', 'LV2', 'LV3', 'LV4']
    loading_df = pd.DataFrame(loadings, index=X.index, columns=LV_list)
    
    ## Save a loadings result file with gene_info, log2FoldChange, loading_df, normalized counts
    log2FoldChange_df = pd.DataFrame(log2FC_calculate(X, class_list))
    log2FoldChange_df = log2FoldChange_df.rename(columns={0:'log2FoldChange'})
    
    merged_df = pd.concat([gene_info, log2FoldChange_df, loading_df, X], axis=1)
    merged_df.to_excel(savedir + '0_all_Loadings.xlsx')
    
    ## Find the major contributors (Top 1000) of each loading variable and save
    a=1
    for i in LV_list:
        # Select genes with weight >= 0.01 as major contributors of each latent variable
        major_merged_df = merged_df[abs(merged_df[i]) >= 0.01]
        # Save all top 1000 genes in the current LV
        major_merged_df.to_excel(savedir + f'{a}_A_' + i + '_ALL_major_contributors.xlsx')
        # Select and save down regulated genes in the current LV
        down_major_merged_df = major_merged_df[major_merged_df['log2FoldChange'] < 0]
        down_major_merged_df.to_excel(savedir + f'{a}_B_' + i + '_DOWN_major_contributors.xlsx')
        # Select and save up regulated genes in the current LV
        up_major_merged_df = major_merged_df[major_merged_df['log2FoldChange'] > 0]
        up_major_merged_df.to_excel(savedir + f'{a}_C_' + i + '_UP_major_contributors.xlsx')
        a+=1
        
    ## Plot with ellipse
    #For plotting, create a list of conditions
    condition_list = metadata_drop[design_factor].tolist()
    # Extracting the scores of the two latent variables
    scores = plsda.x_scores_  
    
    for i in LV_list:
        if i == 'LV1':
            continue
        else:
            plot_PLSDA_with_ellipse(scores = scores,
                                    condition_list = condition_list,
                                    exp_condition = exp_condition,
                                    ref_condition = ref_condition,
                                    LV_list=LV_list,
                                    x_axis_LV='LV1',
                                    y_axis_LV=i,
                                    savedir=savedir)
    return
    
```

```python
# Plot with Ellipse on selected LVs
# The size and orientation of the ellipse are determined by the covariance matrix, and scaled with confidence of 0.95
# Ellipse center is the mean of the group's points

def plot_PLSDA_with_ellipse(scores, condition_list, exp_condition, ref_condition, LV_list, x_axis_LV, y_axis_LV, savedir):
    # Find index of x_axis_LV and y_axis_LV for plotting use
    index_x_axis_LV = LV_list.index(x_axis_LV)
    index_y_axis_LV = LV_list.index(y_axis_LV)
    
    # Group data points by labels
    groups = {}
    for i, label in enumerate(condition_list):
        if label not in groups:
            groups[label] = []
        groups[label].append(scores[i])
    

    import matplotlib.patches as patches
    from scipy.stats import chi2

    plt.figure(figsize=(8, 6))

    # Define a color for each group
    if condition_list[0] == exp_condition:
        group_colors = {exp_condition: 'blue', ref_condition: 'yellow'}
    else:
        group_colors = {exp_condition: 'yellow', ref_condition: 'blue'}


    for label, group in groups.items():
        group = np.array(group)
        group = group[:, [index_x_axis_LV, index_y_axis_LV]]
        color = group_colors[label]
    
        # Scatter plot for the group
        plt.scatter(group[:, 0], group[:, 1], alpha=0.7, label=label)

        # Calculate the mean and covariance
        group_mean = np.mean(group, axis=0)
        group_cov = np.cov(group, rowvar=False)

        # Calculate the width, height, and angle for the ellipse
        lambda_, v = np.linalg.eig(group_cov)
        ## Determine scale for 95% confidence
        k = np.sqrt(chi2.ppf(0.95, 2))
        lambda_2sqrt = 2 * k * np.sqrt(lambda_)
        ell_width, ell_height = lambda_2sqrt[0], lambda_2sqrt[1]
        angle = np.rad2deg(np.arctan2(*v[:,0][::-1]))

        # Create an Ellipse patch
        ell = patches.Ellipse((group_mean[0], group_mean[1]), ell_width, ell_height, angle=angle, edgecolor=color, facecolor=color, alpha=0.3)
    
        # Add the Ellipse to the plot
        plt.gca().add_patch(ell)

    plt.xlabel(x_axis_LV)
    plt.ylabel(y_axis_LV)
    plt.title('PLS-DA: Scatter Plot')
    plt.legend()
    plt.savefig(savedir + x_axis_LV + '_vs_' + y_axis_LV + '_PLSDA_Plot.pdf')
    
    return
```

```python
def quick_paired_PLSDA_StandardScale(counts_df, metadata, gene_info, design_factor, exp_condition, ref_condition, savedir):
    """
    Performs PLSDA after standard scaling
    
    """
    
    # Find the samples that will not be used in this comparison
    drop_indexes = metadata.index[~metadata[design_factor].isin([exp_condition, ref_condition])].tolist()
    
    # Drop the not needed samples from counts_df and metadata
    counts_df_drop = counts_df.drop(drop_indexes, axis = 1)
    metadata_drop = metadata.drop(drop_indexes, axis=0)
    print(metadata_drop)
    
    ## Prepare counts_df_drop for PLS-DA
    selected_sample_list = counts_df_drop.columns #Retrieve a list of counts_df_drop column names
    
    # Annotate ref as class 0 and exp as class 1
    class_list = []
    for i in selected_sample_list:
        if metadata_drop.loc[i,'condition'] == ref_condition:
            class_list.append('0') # ref_condition is class 0
        elif metadata_drop.loc[i,'condition'] == exp_condition:
            class_list.append('1') # exp_condition is class 1
        else:
            print('error: third class exist!')
            return
    
    # Annotate each sample in counts_df_drop
    temp_index = len(counts_df_drop)
    counts_df_drop.loc[temp_index] = class_list
    counts_df_drop = counts_df_drop.rename(index={temp_index: 'condition'})
    
    # use PLSRegression class from scikit-learn to implement PLS-DA (Partial Least Squares Discriminant Analysis)
    from sklearn.cross_decomposition import PLSRegression
    from sklearn.model_selection import train_test_split
    
    # Separate features and labels
    X = counts_df_drop.drop('condition', axis=0)  #Features
    y = counts_df_drop.loc['condition'] #Labels

    print(f'features:{len(X)}; target:{len(y)}')
    
    from sklearn.preprocessing import StandardScaler

    # Scale the features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # PLS-DA 
    plsda = PLSRegression(n_components=4)  # Choose the number of components
    plsda.fit(X_scaled.T, y)
    
    # Extract loadings for each latent variables for future analysis
    loadings = plsda.x_loadings_

    # Convert the loadings for each feature to a dataframe
    LV_list = ['LV1', 'LV2', 'LV3', 'LV4']
    loading_df = pd.DataFrame(loadings, index=X.index, columns=LV_list)
    
    ## Save a loadings result file with gene_info, log2FoldChange, loading_df, normalized counts
    log2FoldChange_df = pd.DataFrame(log2FC_calculate(X, class_list))
    log2FoldChange_df = log2FoldChange_df.rename(columns={0:'log2FoldChange'})
    
    merged_df = pd.concat([gene_info, log2FoldChange_df, loading_df, X], axis=1)
    merged_df.to_excel(savedir + '0_all_Loadings.xlsx')
    
    ## Find the major contributors (Top 1000) of each loading variable and save
    a=1
    for i in LV_list:
        # Select genes with weight >= 0.01 as major contributors of each latent variable
        major_merged_df = merged_df[abs(merged_df[i]) >= 0.01]
        # Save all top 1000 genes in the current LV
        major_merged_df.to_excel(savedir + f'{a}_A_' + i + '_ALL_major_contributors.xlsx')
        # Select and save down regulated genes in the current LV
        down_major_merged_df = major_merged_df[major_merged_df['log2FoldChange'] < 0]
        down_major_merged_df.to_excel(savedir + f'{a}_B_' + i + '_DOWN_major_contributors.xlsx')
        # Select and save up regulated genes in the current LV
        up_major_merged_df = major_merged_df[major_merged_df['log2FoldChange'] > 0]
        up_major_merged_df.to_excel(savedir + f'{a}_C_' + i + '_UP_major_contributors.xlsx')
        a+=1
        
    ## Plot with ellipse
    #For plotting, create a list of conditions
    condition_list = metadata_drop['condition'].tolist()
    # Extracting the scores of the two latent variables
    scores = plsda.x_scores_  
    
    for i in LV_list:
        if i == 'LV1':
            continue
        else:
            plot_PLSDA_with_ellipse(scores = scores,
                                    condition_list = condition_list,
                                    exp_condition = exp_condition,
                                    ref_condition = ref_condition,
                                    LV_list=LV_list,
                                    x_axis_LV='LV1',
                                    y_axis_LV=i,
                                    savedir=savedir)
    return
    
```
